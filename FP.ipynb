```python
import pandas as pd
```

```python
df=pd.read_csv(path,
              header=None)
df
```

```python
ndf=df.applymap(str)
trans=ndf.values.tolist()
trans
```

```python
pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
te_ary=te.fit(trans).transform(trans)
```

```python
encodeDF=pd.DataFrame(te_ary,
                      columns=te.columns_)
encodeDF
```

```python
first10=encodeDF#.head()
ndf=first10.copy()
ndf
```

```python
ndf=ndf.drop('nan',axis=1)
ndf.columns
```

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
fre_itms=fpgrowth(ndf,
                 min_support=0.6,
                 use_colnames=True)
fre_itms
```

```python
as_ruls=association_rules(fre_itms,
                         metric='lift',
                         min_threshold=1.5)
as_ruls
```

```python
## which are the top 10 frequent patterns with high Confidence?
```

```python
sortConf=as_ruls.sort_values('confidence',
                                ascending=False).head(10)
sortConf                                                
```

```python
#Caso # 2 Transactions with antrecedents (m)
```

```python
sortConf[sortConf['antecedents']=={'m'}]
```
